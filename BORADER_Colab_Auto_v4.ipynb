# 🎥 BORADER Video Downloader - Google Colab Version

Download videos from YouTube, TikTok, Instagram, and 1000+ sites directly in Google Colab!

**✅ Works for ALL YouTube videos** - No authentication required!

---

```python
# Install yt-dlp
!pip install yt-dlp

# Download the main downloader script
!wget -q https://raw.githubusercontent.com/bora2025/borader-downloader/main/main.py

# Create downloads directory
!mkdir -p downloads

# Auto-detect URL from browser fragment
from IPython.display import Javascript, HTML
import time

# Initialize URL variable
video_url = None

# JavaScript to extract URL from browser fragment
display(HTML('''
<script>
    // Extract URL from page fragment
    const fragment = window.location.hash.substring(1);
    const params = new URLSearchParams(fragment);
    const url = params.get('video_url');
    
    if (url) {
        // Store URL in global variable for Python to access
        window.detectedVideoUrl = decodeURIComponent(url);
        console.log('🎯 URL detected:', window.detectedVideoUrl);
        
        // Show user feedback
        const statusDiv = document.createElement('div');
        statusDiv.innerHTML = '<p style="color: green; font-weight: bold;">✅ URL detected! Starting download...</p>';
        statusDiv.id = 'urlStatus';
        document.body.appendChild(statusDiv);
    } else {
        console.log('No URL detected in fragment');
    }
</script>
'''))

# Wait for JavaScript to execute
time.sleep(3)

# Check if URL was detected and set Python variable
display(Javascript('''
    if (window.detectedVideoUrl) {
        // Set Python variable directly
        IPython.notebook.kernel.execute('video_url = "' + window.detectedVideoUrl.replace(/\"/g, '\\"') + '"');
    }
'''))

# Additional wait for variable to be set
time.sleep(2)

# Status message
print('✅ Setup complete!')
if video_url:
    print(f'🎯 Auto-detected URL: {video_url}')
    print('🚀 Download will start automatically...')
else:
    print('No URL detected - ready for manual input')
    video_url = 'YOUR_VIDEO_URL'  # Replace with your URL for manual testing
```

## 💾 Connect Google Drive (Recommended)

Mount Google Drive to save videos permanently:

```python
# Mount Google Drive
try:
    from google.colab import drive
    drive.mount('/content/drive')
    
    # Create BORADER downloads folder
    import os
    drive_path = '/content/drive/My Drive/BORADER_Downloads'
    os.makedirs(drive_path, exist_ok=True)
    
    print('✅ Google Drive mounted!')
    print(f'📁 Videos will save to: {drive_path}')
    output_dir = drive_path
    
except Exception as e:
    print(f'❌ Google Drive mount failed: {e}')
    print('💡 Using Colab storage (temporary)')
    output_dir = 'downloads'
```

## 📥 Download Video

The download will start automatically if a URL was detected above!

```python
# Download the video
if video_url and video_url != 'YOUR_VIDEO_URL':
    print(f'🎬 Downloading: {video_url}')
    print(f'💾 Saving to: {output_dir}')
    
    # Execute download
    !python main.py "$video_url" --output "$output_dir"
    
    print('\n✅ Download completed!')
    print(f'📂 Check your files in: {output_dir}')
    
else:
    print('❌ No valid URL found!')
    print('💡 To download manually:')
    print('   1. Set video_url above to your video URL')
    print('   2. Run this cell again')
```
