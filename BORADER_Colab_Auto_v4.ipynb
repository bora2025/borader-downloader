# 🎥 BORADER Video Downloader - Google Colab Version

Download videos from YouTube, TikTok, Instagram, and 1000+ sites directly in Google Colab!

**✅ Works for ALL YouTube videos** - No authentication required!

```python
# Install yt-dlp
!pip install yt-dlp

# Download the main downloader script
!wget -q https://raw.githubusercontent.com/bora2025/borader-downloader/main/main.py

# Create downloads directory
!mkdir -p downloads

# Auto-detect URL from page fragment
from IPython.display import Javascript
import time

detected_url = None
display(Javascript('''
    const fragment = window.location.hash.substring(1);
    const params = new URLSearchParams(fragment);
    const videoUrl = params.get('video_url');
    if (videoUrl) {
        IPython.notebook.kernel.execute('detected_url = "' + decodeURIComponent(videoUrl) + '"');
    }
'''))

time.sleep(2)
print('✅ Setup complete!')
if detected_url:
    print(f'🎯 Auto-detected URL: {detected_url}')
    print('🚀 Starting download...')
    video_url = detected_url
    !python main.py "$video_url" --output downloads
else:
    print('Manual mode: Set video_url below')
```
